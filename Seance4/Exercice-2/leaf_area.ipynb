```python
import os
import matplotlib.pyplot as plt

from skimage.io import imread
from skimage.color import rgb2gray
from skimage.filters import threshold_otsu, threshold_li
from collections import Counter
```

```python
# Input
# une_image: image input
# filter: name of selected filter
selected_image = imread(r"C:\Users\KiloO\Deep-image-processing\Seance4\Exercice-2\PLPATH-FRU-35-Strawberry-Leaf-Diseases-figure-2.jpg")
filter_selected = 'otsu'

# MAIN SCRIPT
gray_ = rgb2gray(selected_image) # rgb image to gray image
if filter_selected == 'otsu':
    ths = threshold_otsu(gray_) # determine threshold with otsu method
elif filter_selected == 'li':
    ths = threshold_li(gray_) # determine threshold with li method
mask = (ths > gray_) # apply the threshold to the image

plt.imshow(mask,cmap="gray") # Visualization
plt.axis('off')
```
